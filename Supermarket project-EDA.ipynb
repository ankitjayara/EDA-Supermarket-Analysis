### Familiarize with the dataset and variables: 

* Invoice ID : computer Generated sale slips invoice identification number.
* Branch:  Branches of supermakter(there are 3 branches identified as A,B,C).
* City:  Location Of supermarket.
* Customer Type:  two Type of Customer Member and Normal.
* Gender:  Gender type customer(male/female)
* Product line:  General item categorization groups .  
* Unit Price:  price of each product.
* Quantity:  Number of product purchased by customer.
* Tax 5% :  5% tax fee.
* Total:  Total price including tax.
* Date:  Date of Purchase.
* Time: Time of Purchase.
* Payment:  Pyment used for Purchase(Ewallet,cash,credit card).
* cogs:  cost of good sold.
* gross margin percentage:Gross margin Percentage.
* gross income:  Gross income from customer.
* Rating: Customer rating on their overall shopping experience.

###  What can you say about the aggregate sales across the branches

```python
t=df.groupby("Branch").agg(aggregate_sales=("Quantity","sum"))
t=t.reset_index()
plt.figure(figsize=(6,6))
sns.barplot(x="Branch",
            y="aggregate_sales",
            data=t)
plt.title('Aggregate Sales accross Branches')
```

### Which is the most profitable branch 

```python
avg=df.groupby("Branch")["gross income"].mean()
x=list(avg.index)
y=list(avg.values)
plt.xlabel('Branches')
plt.ylabel('Profit')
plt.title('most profitable Branch')
plt.bar(x,y,color=['red','blue','cyan'])
```

###  Which product line helps you generate the most income 

```python
avg=df.groupby('Product line')['gross income'].mean()
plt.figure(figsize=(20,8))
x=list(avg.values)
y=list(avg.index)
plt.xlabel('Product line')
plt.ylabel('Gross Income')
plt.title('Most Income Products')
plt.bar(y,x,color=['red','blue','cyan','green','orange','green'])
```

```python
# As we have to find the most income so we have to find the sum of gross income for each product line
#rather then mean of gross income.
```

```python
total=df[['Product line','gross income']].groupby(['Product line'],
                                                  as_index=False).sum().sort_values(by='gross income',ascending=False)
plt.figure(figsize=(15,10))
sns.barplot(x='Product line',y='gross income',data=total)
```

### so food and beverages ,sports and travel generate most income 

### What is the spending pattern of both males and females, meaning in which category do they spend more 

```python
t=df.groupby(["Product line","Gender"]).agg(mean_total=("Total","mean"))
t=t.reset_index()
plt.figure(figsize=(20,10))
sns.barplot(x="Product line",y="mean_total",hue="Gender",data=t)
```

#### Male category spends more in Health and Beauty whereas Female spends more in Home and lifestyle
